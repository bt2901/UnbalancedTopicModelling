```python
DATA_PATH = '../data/lenta_1000_100_short.csv'

dataset_lenta = Dataset(DATA_PATH, batch_size=2000)
dictionary_lenta = dataset_lenta.get_dictionary()
batch_vectorizer_lenta = dataset_lenta.get_batch_vectorizer()
```
